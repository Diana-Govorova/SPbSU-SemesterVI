```python
import numpy as np
from numpy import linalg as li
```

Степенной метод

```python
def degree(A, e):
    x = np.ones_like(A[0])
    cond = True
    l = 0
    count = 1
    x_new = np.matmul(A,x)
    l_new = x_new / x

    while np.all(l_new - l) > e:
        x = x_new
        l = l_new
        count+=1
        x_new = np.matmul(A,x)
        l_new =  x_new / x
    vector = np.split(x, len(x))
    return count, abs(l[0]), vector / vector[len(x) - 1]
```

Метод скалярных произведений

```python
def scalar(A, e):
    At = A.transpose()
    x = np.ones_like(A[0])
    count = 1
    X = np.dot(A, x)
    Y = np.dot(At, x)
    Z = np.dot(li.matrix_power(A, 0), X)
    l = 0
    l_new = np.dot(X, Y) / np.dot(Z, Y)
    
    while np.all(l_new - l) > e:
        l = l_new
        Z = X
        X = np.dot(A, X)
        Y = np.dot(At, Y)
        l_new = np.dot(X, Y) / np.dot(Z, Y)
        count+=1
    vector = np.split(Z, len(Z))    
    return count, abs(l), vector / vector[len(Z) - 1]   
        
    
```
